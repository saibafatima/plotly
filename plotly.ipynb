```python
#Q1: Load the "titanic" dataset and plot a scatter plot using Plotly Express

import seaborn as sns
import plotly.express as px

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Plot a scatter plot
fig = px.scatter(titanic, x="age", y="fare")
fig.update_layout(title="Age vs Fare in Titanic Dataset")
fig.show()
```

```python
#Q2: Plot a box plot using the "tips" dataset and Plotly Express

import plotly.data as pdata
import plotly.express as px

# Load the tips dataset
tips = pdata.tips()

# Plot a box plot
fig = px.box(tips, x="day", y="total_bill")
fig.update_layout(title="Total Bill by Day of the Week")
fig.show()
```

```python
#Q3: Plot a histogram using the "tips" dataset and Plotly Express
import plotly.data as pdata
import plotly.express as px

# Load the tips dataset
tips = pdata.tips()

# Plot a histogram
fig = px.histogram(tips, x="total_bill", color="sex", pattern_shape="smoker", facet_col="day")
fig.update_layout(title="Total Bill Distribution by Sex, Smoker, and Day")
fig.show()
```
